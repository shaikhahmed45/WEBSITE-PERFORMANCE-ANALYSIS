WEBSITE - DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('website.csv')
```

```python
df
```

```python
df.columns = df.iloc[0]
df
```

```python
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ["channel group","DateHour","Users","Seession","Engaged Sessions","Average engagement time per session","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df
```

```python
df.info()
```

```python
df["DateHour"] = pd.to_datetime(df["DateHour"],format="%Y%m%d%H", errors = 'coerce')
df.head(3)
```

```python
numeric_cols = df.columns.drop(["channel group","DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric,errors='coerce')
df["Hour"] = df["DateHour"].dt.hour
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

SESSIONS ANS USER USER OVERTIME

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Seession","Users"]].sum().plot(ax=plt.gca())
plt.title("SEESSION AND USER OVER TIME")
plt.xlabel("DateHour")
plt.ylabel("count")
plt.show()
```

TOTAL USERS BY CHANNEL

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="channel group",y="Users",estimator=np.sum,palette="viridis")
plt.title("TOTAL USERS BY CHANNEL")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

AVERAGE ENGAGEMENT TIME BY CHANNELS

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x="channel group",y="Average engagement time per session",estimator=np.mean,palette="magma")
plt.title("AVG ENGAGEMENT TIME BY CHANNEL")
plt.xticks(rotation=45)
plt.show()
```

ENGAGEMENT RATE DISTRIBUTION BY CHANNEL

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df,x="channel group",y="Engagement rate",palette="coolwarm")
plt.title("ENGAGEMENT RATE DISTRIBUTION BY CHANNEL")
plt.xticks(rotation=45)
plt.show()
```

ENGAGED VS NON ENGAGED SEESSIONS

```python
session_df = df.groupby("channel group")[["Seession","Engaged Sessions"]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Seession"] - session_df["Engaged Sessions"]
session_df_melted = session_df.melt(id_vars="channel group",value_vars=["Engaged Sessions","Non-Engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted,x="channel group",y="value",hue="variable")
plt.title("ENGAGED VS NON-ENGAGED")
plt.xticks(rotation=45)
plt.show()
```

TRAFFIC BY HOURS AND CHANNEL

```python
heatmap_data = df.groupby(["Hour","channel group"])["Seession"].sum().unstack().fillna(0)

plt.figure(figsize=(8,5))
sns.heatmap(heatmap_data, cmap="YlGnBu", linewidth=.5 , annot=True , fmt = '.0f')
plt.title("TRAFFIC BY HOUR AND CHANNEL")
plt.xlabel("Channel Group")
plt.ylabel("Hour of Day")
plt.show()
```

ENGAGEMENT RATE VS SEESSIONS

```python
df.head(2)
```

```python
df_plot = df.groupby("DateHour")[["Engagement rate","Seession"]].mean().reset_index()

plt.figure(figsize=(8,5))
plt.plot(df_plot["DateHour"], df_plot["Engagement rate"], label="Engagement rate", color="red")
plt.plot(df_plot["DateHour"], df_plot["Seession"], label="Seessions", color="green")
plt.title("Engagement rate vs Seessions overtime")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
